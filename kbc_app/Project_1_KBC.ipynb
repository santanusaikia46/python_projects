```python
import random

# Define a list to store questions and their associated data
question_bank = [
    {
        "question": "What is the capital of India?",
        "options": ["New Delhi", "Mumbai", "Kolkata", "Chennai"],
        "correct_answer": "New Delhi",
    },
    {
        "question": "In Python, what is the result of 3 + 4?",
        "options": ["5", "6", "7", "8"],
        "correct_answer": "7",
    },
    {
        "question": "Which of the following is a programming language?",
        "options": ["Java", "Lion", "Elephant", "Giraffe"],
        "correct_answer": "Java",
    },
    {
        "question": "What is the chemical symbol for water?",
        "options": ["H2O", "CO2", "O2", "NaCl"],
        "correct_answer": "H2O",
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "options": ["Mars", "Venus", "Jupiter", "Saturn"],
        "correct_answer": "Mars",
    },
    {
        "question": "In Python, which keyword is used to define a function?",
        "options": ["def", "function", "define", "func"],
        "correct_answer": "def",
    },
    {
        "question": "What is the largest mammal on Earth?",
        "options": ["Blue Whale", "Elephant", "Giraffe", "Lion"],
        "correct_answer": "Blue Whale",
    },
    {
        "question": "Which gas do plants use for photosynthesis?",
        "options": ["Oxygen", "Carbon Dioxide", "Hydrogen", "Nitrogen"],
        "correct_answer": "Carbon Dioxide",
    },
    {
        "question": "What is the largest planet in our solar system?",
        "options": ["Jupiter", "Saturn", "Earth", "Mars"],
        "correct_answer": "Jupiter",
    },
    {
        "question": "In Python, which data type is used to store a single character?",
        "options": ["char", "string", "character", "single"],
        "correct_answer": "string",
    },
    {
        "question": "What is the chemical symbol for oxygen?",
        "options": ["O2", "H2O", "CO2", "N2"],
        "correct_answer": "O2",
    },
    {
        "question": "Who is known as the Father of the Indian Constitution?",
        "options": ["B.R. Ambedkar", "Mahatma Gandhi", "Jawaharlal Nehru", "Sardar Patel"],
        "correct_answer": "B.R. Ambedkar",
    },
    {
        "question": "What is the primary function of the heart in the human body?",
        "options": ["Pumping blood", "Digesting food", "Filtering air", "Producing hormones"],
        "correct_answer": "Pumping blood",
    },
    {
        "question": "Which gas is responsible for the blue color of the Earth's sky?",
        "options": ["Nitrogen", "Oxygen", "Carbon Dioxide", "Helium"],
        "correct_answer": "Nitrogen",
    },
    {
        "question": "What is the national flower of India?",
        "options": ["Lotus", "Rose", "Tulip", "Sunflower"],
        "correct_answer": "Lotus",
    },
    {
        "question": "In Python, what does the `print` function do?",
        "options": ["Displays text on the screen", "Calculates math expressions", "Plays music", "Takes photos"],
        "correct_answer": "Displays text on the screen",
    }
]

# Define a function to display questions
def display_question(question_data):
    print(question_data["question"])
    for i, option in enumerate(question_data["options"], start=1):
        print(f"{i}. {option}")

# Define a 50-50 lifeline function
def fifty_fifty(question_data):
    # Create a copy of the options list
    options_copy = question_data["options"][:]

    # Remove two incorrect options randomly
    while len(question_data["options"]) > 2:
        incorrect_option = random.choice(options_copy)
        if incorrect_option != question_data["correct_answer"]:
            question_data["options"].remove(incorrect_option)
            options_copy.remove(incorrect_option)

    # Display the updated question with only two options remaining
    print(question_data["question"])
    for i, option in enumerate(question_data["options"], start=1):
        print(f"{i}. {option}")

# Define a list of level winnings
level_winnings = [1000, 2000, 3000, 4000, 5000, 10000, 20000, 40000, 80000, 160000, 320000, 640000, 1250000, 2500000, 500000, 1000000]

# Define a function to run the game
def game():
    print("Welcome to 'Kaun Banega Crorepati'!")

    total_winnings = 0  # Initialize total winnings
    level = 0  # Initialize the level
    lifelines = {"50-50": True}  # Initialize lifelines (50-50 lifeline available at the start)

    # Shuffle the question bank to present questions in random order
    random.shuffle(question_bank)

    for question_data in question_bank:
        display_question(question_data)

        if lifelines["50-50"]:
            print("Lifelines available: 50-50")
            lifeline_choice = input("Do you want to use a lifeline? (yes/no): ").strip().lower()

            if lifeline_choice == "yes":
                fifty_fifty(question_data)
                lifelines["50-50"] = False  # Mark 50-50 lifeline as used
            elif lifeline_choice == "no":
                pass  # Continue without using a lifeline
            else:
                print("Invalid input. Continuing without using a lifeline.")

        # Get user's answer choice
        user_choice = int(input("Enter the number of option: "))

        # Validate user input
        if user_choice < 1 or user_choice > 4:
            print("Invalid choice. Please enter a valid choice.")
            continue

        selected_option = question_data["options"][user_choice - 1]

        if selected_option == question_data["correct_answer"]:
            total_winnings = level_winnings[level]
            print("Correct answer! You won", total_winnings, "points.\n")
            level += 1
        else:
            print("Sorry, that's incorrect. The correct answer was:", question_data["correct_answer"], "\n")
            break

    print("Congratulations! You won a total of", total_winnings, "points.")
    print("Thank you for playing!")

# Entry point of the program
if __name__ == "__main__":
    game()
```
